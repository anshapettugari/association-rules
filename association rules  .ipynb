```python
import numpy as np
import pandas as pd
```

```python
arules=pd.read_csv("book.csv")
arules
```

```python
!pip install mlxtend
```

```python
from mlxtend.frequent_patterns import apriori,association_rules
frequent_books = apriori(arules, min_support=0.05,max_len=5,use_colnames = True)
frequent_books
```

```python
frequent_books1 = apriori(arules, min_support=0.005,max_len=3,use_colnames = True)
frequent_books1
```

```python
frequent_books.sort_values('support',ascending = False,inplace=True)
frequent_books.sort_values 
```

```python
rules = association_rules(frequent_books, metric="lift", min_threshold=1)
rules.head(20)
rules.sort_values('lift',ascending = False).head(10)  
```

```python
rules = association_rules(frequent_books1, metric="lift", min_threshold=1)
rules.head(20)
rules.sort_values('lift',ascending = False).head(10)  
```

```python
amovies=pd.read_csv("my_movies.csv")
amovies
```

```python
amovies.describe()
```

```python
amovies1=amovies.drop(['V1','V2','V3','V4','V5'],axis=1)
amovies1
```

```python
amovies1.info()
```

```python
from mlxtend.frequent_patterns import apriori,association_rules
frequent_movies = apriori(amovies1, min_support=0.005,max_len=4,use_colnames = True)
frequent_movies
```

```python
frequent_movies.sort_values('support',ascending = False,inplace=True)
frequent_movies.sort_values 
```

```python
rules = association_rules(frequent_movies, metric="lift", min_threshold=1)
rules.head(20)
rules.sort_values('lift',ascending = False).head(10)  
```
